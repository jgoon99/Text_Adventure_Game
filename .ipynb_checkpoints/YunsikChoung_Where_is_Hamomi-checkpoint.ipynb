```python
"""
=========================================================

Title: Where's Halmoni

=========================================================
"""
# Importing Packages
import random # Make a game to randomize
import math # Calculate playing time
import time # Calculate  playing time
import sys
from IPython.display import clear_output
from typing import Optional


# START OF FUNCTIONAL METHOD ################################################################################## #
# Slow print and clear console
def sprint(msg, cls: Optional[bool] = False):
    """
    Method Name: sprint
    Arguments: msg
    Description: Print slowly
    """
    # slowly print timer
    slowlyTime = 0.001
    if cls:
        clear_output(wait=False)
    for c in list(msg):
        sys.stdout.write(c)
        sys.stdout.flush()
        time.sleep(slowlyTime)

# startGame Methode
def startGame():
    """
    MethodName: startGame
    Description: Main Method of game.
                 Run startGame will start the game
    """
    # Save the starting Time
    startTime = time.time()
    
    # The Book Introduction
    bookIntroduce()
    
    ## GAME OVER CONTROL
    gamePass = True # if GameOver TRUE will Shut Down.
    
    # Declare Pass variable
    stage = 1
    
    # challenge counter
    challengeCounter = 1
    while True:
        
        if gamePass: # User is playing game
            if stage == 1:
                gamePass = chapterOne()
                if gamePass:
                    # Get Script
                    getChapterOneClearSciript()
                    stage = 2
                else:
                    getChapterOneFailScript()
            elif stage == 2:
                gamePass = chapterTwo()
                if gamePass:
                    # Get Script
                    getChapterTwoClearSciript()
                    stage = 3
                else:
                    getChapterTwoFailScript()
            elif stage == 3:
                gamePass = chapterThree()
                if gamePass:
                    getChapterThreeClearSciript()
                    stage = 4
                else:
                    getChapterThreeFailScript()
            elif stage == 4:
                gamePass = chapterFour()
                if gamePass:
                    getChapterFourClearSciript()
                    stage = 5
                    break
                else:
                    getChapterFourFailScript()
            else:
                print("Unexpacted Value on Stage " + str(stage))
        else: # User fail the game
            print(f"""You failed to clear the Chapter {stage}!\n\nDo you want to try again?\n""")
            keepGoing = input("""Please enter <Yes/No> to choose an option.\n\n""")
            if keepGoing != '' and (keepGoing.lower() == 'y' or keepGoing == 'yes' or keepGoing == 'ok'):
                gamePass = True
                continue
            elif keepGoing != '' and (keepGoing.lower() == 'n' or keepGoing == 'no' or keepGoing == 'nope'):
                print("""Thanks for playing Where's Halmoni?\n\n""")
                break
            else:
                print("Please enter Yes or No!")
    # Calulating Playing time
    endTime = time.time()
    gap = endTime - startTime
    min = math.floor(gap/60)
    sec = math.floor(gap%60)
    gap_time = ''
    if min < 10:
        gap_time += '0' + str(min)
    else:
        gap_time += str(min)
    gap_time += ':'
    if sec < 10:
        gap_time += '0' + str(sec)
    else:
        gap_time += str(sec)
    
    print(f"""\n\nYou played : {gap_time}\n\n\n\n""")
    input(prompt="Press <Enter> to Resume\n\n")
    if gamePass:
        win()
    else:
        fail()
    
# END ##################### startGame ###############################################
    
# ChapterOne Method
def chapterOne():
    """
    Method: chaterOne
    Return: response [Boolean]: True = pass the Chapter one, False = Game over!
    Description: Exploring children's house with baseball game.
                 Guessing 3 Digit Passward.
                 Default PWD = 356
                 User can find some hints in living and ketchen.
                 First digit: Number of Masks on the wall in living room
                 Second digit: Broken Timer's minute first digit in the kitchen
                 Third digit: Broken Timer's minute second digit in the kitchen
    """
    # Declare Response
    response = False
    
    # Get Chapter One Starting Script
    getChapterOneStartScript()
    
    # Declear Position: 
    position = 'foyer' # First Position
    
    # action counting
    actionCount = 0
    
    # Declear While Controler
    explore = True # When user exploring, explore should be True
    
    # Joon's hint list
    hintList = [
        'Joon said, "Noona, I think something new in Halmoni\'s Bedroom, Let\'s go and investigate that room!"\n',
        'Joon said, "Noona, I saw some masks on the wall in the living room. Is that any meaningful?"\\n',
        'Joon said, "Noona, Halmoni\'s kitchen timer has broken in the kitchen. What time was that?\n"',
        'Joon said, "Three masks and five..fif...fifteen? fifty something? It\'s a little bit hard to read a big number for me."\n'
    ]
    
    # Explore While Loop
    while explore:
        # add 1 actionCount
        actionCount += 1
        
        # if some user got lost in the house, give some hine
        if actionCount > 8 and position != 'baseball':
            print(random.choice(seq = hintList))
        else :
            pass 
        
        # Calling each place function after check position
        if position == 'foyer': 
            position = foyer()
        elif position == 'living':
            position = living()
        elif position == 'bathroom':
            position = bathroom()
        elif position == 'kitchen':
            position = kitchen()
        elif position == 'bedroom':
            position = bedroom()
        elif position == 'baseball':
            # Break Explore While
            # position baseball means that to start baseball game
            break
        else: # Unexpected Position
            print('Error - Unexpected position: ' + position)
            print('On Explore While in Chapter 1')
            # Debuging Break
            break
    # END OF explore WHILE 
    
    # BASEBALL GAME CHECK
    if position == 'baseball':
        print(f"""
==============================================================================
  ____                 _           _ _    _____                      _ _ _ 
 |  _ \               | |         | | |  / ____|                    | | | |
 | |_) | __ _ ___  ___| |__   __ _| | | | |  __  __ _ _ __ ___   ___| | | |
 |  _ < / _` / __|/ _ \ '_ \ / _` | | | | | |_ |/ _` | '_ ` _ \ / _ \ | | |
 | |_) | (_| \__ \  __/ |_) | (_| | | | | |__| | (_| | | | | | |  __/_|_|_|
 |____/ \__,_|___/\___|_.__/ \__,_|_|_|  \_____|\__,_|_| |_| |_|\___(_|_|_)
                                                                           
==============================================================================

!!!!!!!!!!!!!!!!!   NOW YOU WILL PLAY BASEBALL GAME  !!!!!!!!!!!!!!!!!!!!!

        """, True)
        chance = 5 # User have Total 5 chance to guess the password
        input('Press <Enter> to Resume.\n\n')
        
        # Baseball game Trying While
        while chance > 0:
            sprint(f"""
{'-' * 20}
You have {chance} chances to try to guess the password!!
{'-' * 20}

The password is THREE digits.

When you put three number, the magic combination lock would inform you the result.

REMEMBER! Strike means SAME NUMBER ON SAME POSITION,

Ball means SAME NUMBER BUT DIFFERENCT POSITION.

Are you ready to Play?
            
            """, True)
            
            # CALL baseball Function
            result = baseball()
            
            # Check Result Status 1 mean OK, 0 means Error
            if result['status'] == 1: # It's OK
                sprint(f"""You are clicking number blocks.\n""")
                sprint(f"""
[{result['userList'][0]}].......[{result['userList'][1]}].............[{result['userList'][2]}]
.
.
.
                """)
                # Print game result
                sprint(result['msg'] + '\n')
                input('Press <Enter> to Resume.\n\n')
                
                # Checking win or loose
                if result['win']:
                    # user winning result insert to response
                    response = True
                    # break the while User is won.
                    break
                else: # User answer is wrong
                    chance -= 1 # Chance substitude
                    continue
            else:
                sprint(result['msg'] + '\n')
        # End of Baseball game Trying While
    else:
        print('Error - Unexpected position: ' + position)
        print("Not expected Value on Position. On ELSE: baseball check")
    
    # Return response
    return response
# End of chapterOne -------------------------------------------------

# foyer Methond for Stage one.
def foyer():
    """
    Method: foyer
    Return: response[str] - next location string
    """
    # Declare Response 
    response = 'foyer'
    
    # Option Description
    sprint(f"""
Now you are in the foyer. 

What do you want to do?

1. Go forward to other room.
2. Investigate This room.
    
    """, True)
    userInput = input(prompt='Please Select your option.\n\n')
    
    # User Input Verification
    # User select 1
    if userInput != '' and (userInput == '1' or userInput.lower() in list('go forward other')):
        
        # Go forward to other room While
        while True:
            
            # Option descript
            sprint(f"""
Where do you want to go?

1. Living Room
2. Bathroom
3. Kitchen
4. Bedroom.

            """, True)
            userInput = input(prompt = 'please Select your option.\n\n')
            
            # input verification
            if userInput != '' and (userInput == '1' or userInput.lower() in 'living rooom'):
                response = 'living'
                print("""
Children are going forward to the Living room.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '2' or userInput.lower() in 'bathroom'):
                response = 'bathroom'
                print("""
Children are going forward to the Barhroom.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '3' or userInput.lower() in 'kitchen'):
                response = 'kitchen'
                print("""
Children are going forward to the kitchen.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '4' or userInput.lower() in 'bedroom'):
                response = 'bedroom'
                print("""
Children are going forward to the Bedroom.
                
                """)
                # Breaking while, user wants to go to other room
                break
            else: 
                print("""Cannot understand your commend. Please Try it again!""")
        # End of Go forward to other room While
    
    # User select 2 - investigation room
    elif userInput != '' and (userInput == '2' or userInput.lower() in list('investigating this')):
        sprint("""
There are nothing specials in the foyer. Halmoni's shoes are here.

Where is Halmoni?
        
        """, True)
        input('Press <Enter> to Resume.\n\n')
    else: 
        # Nothing Changed And return to ChapterOne
        print("""\nCannot understand your commend. Please Try it again!\n""")
    # Return Response
    return response
# End of foyer -------------------------------------------------     
        
# Method living
def living():
    """
    Method: living
    Return: str - next location string
    """
    response = 'living'
    sprint(f"""
Now you are in the Living Room. 

What do you want to do?

1. Go forward to other room.
2. Investigate This room.
    
    """, True)
    userInput = input(prompt='Please Select your option.\n\n')
    if userInput != '' and (userInput == '1' or userInput.lower() in list('go forward other')):
        while True:
            sprint(f"""
Where do you want to go?

1. Foyer
2. Bathroom
3. Kitchen
4. Bedroom.

            """, True)
            userInput = input(prompt = 'please Select your option.\n\n')

            if userInput != '' and (userInput == '1' or userInput.lower() in 'foyer'):
                response = 'foyer'
                print("""
Children are going forward to the Foyer.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '2' or userInput.lower() in 'bathroom'):
                response = 'bathroom'
                print("""
Children are going forward to the Bathroom.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '3' or userInput.lower() in 'kitchen'):
                response = 'kitchen'
                print("""
Children are going forward to the kitchen.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '4' or userInput.lower() in 'bedroom'):
                response = 'bedroom'
                print("""
Children are going forward to the Bedroom.
                
                """)
                # Breaking while, user wants to go to other room
                break
            else: 
                print("""Cannot understand your commend. Please Try it again!""")
    elif userInput != '' and (userInput == '2' or userInput.lower() in list('investigating this')):
        
        # LIVING ROOM HINTS - '3' Masks is first digit of password
        sprint("""
There are '3' Masks on the wall.

Halmoni is from South Korea. These masks are Korean traditional mask.

One looks like pretty woman, it called Gak-Shi-Tal, which means brand new bride.

Second looks like funny man, it called Hal-Mee-Tal, which means big smiling grand mother.

Last looks like scary monster, it called Dok-Kke-Bee-Tal, which means mischievous goblin.
        
        """, True)
        input('Press <Enter> to Resume.\n\n')
    else: 
        print("""\nCannot understand your commend. Please Try it again!\n""")
    return response
# End of living ------------------------------------------------- 

# Method bathroom
def bathroom():
    """
    Method: bathroom
    Return: str - next location string
    """
    response = 'bathroom'
    sprint(f"""
Now you are in the Bathroom. 

What do you want to do?

1. Go forward to other room.
2. Investigate This room.
    
    """, True)
    userInput = input(prompt='Please Select your option.\n')
    if userInput != '' and (userInput == '1' or userInput.lower() in list('go forward other')):
        while True:
            sprint(f"""
Where do you want to go?

1. Foyer
2. Living Room
3. Kitchen
4. Bedroom.

            """, True)
            userInput = input(prompt = 'please Select your option.\n')

            if userInput != '' and (userInput == '1' or userInput.lower() in 'foyer'):
                response = 'foyer'
                print("""
Children are going forward to the Foyer.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '2' or userInput.lower() in 'linving rooom'):
                response = 'living'
                print("""
Children are going forward to the Living room.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '3' or userInput.lower() in 'kitchen'):
                response = 'kitchen'
                print("""   
Children are going forward to the kitchen.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '4' or userInput.lower() in 'bedroom'):
                response = 'bedroom'
                print("""    
Children are going forward to the Bedroom.
                
                """)
                # Breaking while, user wants to go to other room
                break
            else: 
                print("""Cannot understand your commend. Please Try it again!""")
    elif userInput != '' and (userInput == '2' or userInput.lower() in list('investigating this')):
        sprint("""
There are nothing special in the bathroom.

It's clean and frash. 

Where's Halmoni?

        """, True)
        input('Press <Enter> to Resume.\n\n')
    else: 
        print("""\nCannot understand your commend. Please Try it again!\n""")
    return response
# End of bathroom -------------------------------------------------

# Method Kitchen
def kitchen():
    """
    Method: kitchen
    Return: str - next location string
    """
    response = 'kitchen'
    sprint(f"""
Now you are in the Kitchen. 

What do you want to do?

1. Go forward to other room.
2. Investigate This room.
    
    """, True)
    userInput = input(prompt='Please Select your option.\n\n')
    if userInput != '' and (userInput == '1' or userInput.lower() in list('go forward other')):
        while True:
            sprint(f"""
Where do you want to go?

1. Foyer
2. Living Room
3. Bathroom
4. Bedroom.

            """, True)
            userInput = input(prompt = 'please Select your option.\n')

            if userInput != '' and (userInput == '1' or userInput.lower() in 'foyer'):
                response = 'foyer'
                print("""
Children are going forward to the Foyer.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '2' or userInput.lower() in 'linving rooom'):
                response = 'living'
                print("""
Children are going forward to the Living room.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '3' or userInput.lower() in 'bathroom'):
                response = 'bathroom'
                print("""
Children are going forward to the Bathroom.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '4' or userInput.lower() in 'bedroom'):
                response = 'bedroom'
                print(""" 
Children are going forward to the Bedroom.
                
                """)
                # Breaking while, user wants to go to other room
                break
            else: 
                print("""Cannot understand your commend. Please Try it again!""")
    elif userInput != '' and (userInput == '2' or userInput.lower() in list('investigating this')):
        
        # Kitchen Hints - Timer's minutes are second and third digit of password
        sprint("""
Look! The kitchen is so mess!!

There are lot of foot prints on the floor.

What's that foot prints? Where these are from?

And Halmoni's cooking timer is broken. It pointed [AM 08:56]. 

What happend here? And where is Halmoni?
        
        """, True)
        input('Press <Enter> to Resume.\n\n')
    else: 
        print("""\nCannot understand your commend. Please Try it again!\n""")           
    return response
# End of kitchen -------------------------------------------------

# Method bedroom
def bedroom():
    """
    Method: bedroom
    Return: str - next location string
    """
    response = 'bedroom'
    sprint(f"""
Now you are in the Halmoni's Bedroom. 

What do you want to do?

1. Go forward to other room.
2. Investigate This room.
    
    """, True)
    userInput = input(prompt='Please Select your option.\n\n')
    if userInput != '' and (userInput == '1' or userInput.lower() in list('go forward other')):
        while True:
            sprint(f"""
Where do you want to go?

1. Foyer
2. Living Room
3. Bathroom
4. Kitchen.

            """, True)
            userInput = input(prompt = 'please Select your option.\n\n')

            if userInput != '' and (userInput == '1' or userInput.lower() in 'foyer'):
                response = 'foyer'
                print("""   
Children are going forward to the Foyer.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '2' or userInput.lower() in 'linving rooom'):
                response = 'living'
                print("""
Children are going forward to the Living room.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '3' or userInput.lower() in 'bathroom'):
                response = 'bathroom'
                print("""
Children are going forward to the Bathroom.
                
                """)
                # Breaking while, user wants to go to other room
                break
            elif userInput != '' and (userInput == '4' or userInput.lower() in 'kitchen'):
                response = 'kitchen'
                print("""
Children are going forward to the Kitchen.
                
                """)
                # Breaking while, user wants to go to other room
                break
            else: 
                print("""Cannot understand your commend. Please Try it again!""")
    elif userInput != '' and (userInput == '2' or userInput.lower() in list('investigating this')):
        
        # Bedroom Hint - When user investigating bedroom, it gives user an option, start or not game.
        sprint("""
Look! That's new!! Something big white window is on the wall. It looks like korean traditional.

Oh! There is tiny Combination lock! It has three digits of number! Also, there are some tiny words.

"If you failed to get password in FIVE TIMES, it will be broken!!"

*** Password's hint are scattered across in children's home ***

Do you want to try this?
        
        """, True)
        userInput = input(prompt='Please answer Yes or No\n\n')
        
        if userInput != '' and (userInput.lower() == 'y' or userInput.lower() == 'yes' or userInput.lower() == 'ok'):
            
            # Start baseball password guessing game
            response = 'baseball'
        else:
            # Go back to previous prompt
            response = 'bedroom'
    else: 
        print("""\nCannot understand your commend. Please Try it again!\n""")
    return response
# End of bedroom -------------------------------------------------

# Method Baseball
def baseball():
    """
    Method Name: baseball
    return: object
        response = {
            'status': 0, # 0: error occured 1: run successfuly
            'msg': 'Message of result', # Response Message
            'win': False # 3 strike is Winning
        }
    Description: Check 3 digits and return result as a baseball game
                - Same position & same Number = Strike
                - Same Number but Other position = Ball
                eg. 250 for PWD 279 : 1 strike.
                    270 for PWD 279 : 2 strike.
                    720 for PWD 279 : 2 Ball
    """
    string = input(prompt = 'Please insert 3 digits in number. (000 ~ 999)\n')
    # Build Response Object
    response = {
        'status': 0, # 0: error occured 1: run successfuly
        'msg': 'Message of result', # Response Message
        'win': False, # 3 strike is Winning
        'userList': [] # User's Pwd
    }
    # Correct Pawssword
    pwd = [3, 5, 6]
    # Length of string
    x = len(string)
    if x == 3: # Check if it has 3 digits
        userList = []
        for l in list(string):
            try: # Try to convert each digit to integer
                i = int(l)
                userList.append(i)
            except: # If there is character, catch the error
                response['msg'] = 'Please insert 3 digits in number. (000 ~ 999)'
        uSize = len(userList) # Check 3 digits Integer list Size
        if uSize == 3: # Calculating Score
            response['userList'] = userList
            strike = 0
            ball = 0
            for i, k in enumerate(userList): 
                if k == pwd[i]: # Same position & Same Value is STRIKE
                    strike += 1
                else:
                    for j, o in enumerate(pwd):
                        if i != j:
                            if k == o: # Same Value But Other Position is BALL
                                ball += 1
            if strike == 3: # 3 Strike means that user input vlaue is equal to Password
                response['win'] = True
                response['status'] = 1
                response['msg'] = 'You got it! Password is correct!'
            else: # Response result
                response['status'] = 1
                response['msg'] = str(strike) + ' Strike & ' + str(ball) + ' Ball'
        else:
            print("Unexpected entry. Please check the code for bugs.")
    else: # if it is not 3 digits. 
        response['msg'] = 'Please insert 3 digits password.'
    return response
# End of baseball game -------------------------------------------------

# Method Chapter Two
def chapterTwo():
    """
    Method Name: chapterTwo
    Return: response[boolean] 
    Description: Chapter two - Zodiac Riddle and dice rolling Game!
                 1st: Rat
                 2nd: Ox
                 3rd: Tiger
                 4th: Rabbit
                 5th: Dragon
                 6th: Snake
                 7th: Horse
                 8th: Goat
                 9th: Monkey
                 10th: Rooster
                 11th: Dog
                 12th: Pig
    """
    response = False # Chater two win or fail
    # Zodiac list of dictionaries
    zodiac = [
        {
            'name': 'Rat',
            'question': 'This is the "first" of the repeating 12-year cycle of animals which appear in the East-Asia zodiac. '
        },
        {
            'name': 'Ox',
            'question': 'This  is the "second" of the 12-year periodic sequence (cycle) of animals which appear in the East-Asia zodiac related to the East-Asia calendar, and also appears in related calendar systems.',
        },
        {
            'name': 'Tiger',
            'question': 'This is the "third" of the 12-year cycle of animals which appear in the East-Asia zodiac related to the East-Asia calendar.',
        },
        {
            'name': 'Rabbit',
            'question': 'This is the "fourth" in the twelve-year cycle of animals that appear in the ChiEast-Asianese zodiac related to the East-Asia calendar.',
        },
        {
            'name': 'Dragon',
            'question': 'This is the "fifth" of the 12-year cycle of animals which appear in the East-Asia zodiac related to the East-Asia calendar.',
        },
        {
            'name': 'Snake',
            'question': 'This is the "sixth" of the twelve-year cycle of animals which appear in the ChiEast-Asianese zodiac related to the East-Asia calendar. ',
        },
        {
            'name': 'Horse',
            'question': 'This is the "seventh" of the 12-year cycle of animals which appear in the East-Asia zodiac related to the East-Asia calendar.',
        },
        {
            'name': 'Goat',
            'question': 'This is the "eighth" of the 12-year cycle of animals which appear in the East-Asia zodiac related to the East-Asia calendar. ',
        },
        {
            'name': 'Monkey',
            'question': 'This  is the "ninth" of the 12-year cycle of animals which appear in the East-Asia zodiac related to the East-Asia calendar.',
        },
        {
            'name': 'Rooster',
            'question': 'This is the "tenth" of the 12-year cycle of animals which appear in the East-Asia zodiac related to the East-Asia calendar.',
        },
        {
            'name': 'Dog',
            'question': 'This is "eleventh" of the 12-year cycle of animals which appear in the East-Asia zodiac related to the East-Asia calendar. ',
        },
        {
            'name': 'Pig',
            'question': 'This is the "twelfth" of the 12-year cycle of animals which appear in Chinese zodiac,',
        },
    ]
    zodiac_cp = zodiac.copy() # Copy zodiac Lists Keep origin for reset
    random.shuffle(x = zodiac_cp) # Shuffle zodiac_cp for Chapter 2.
    
    ## Description of Chapter 2.1
    getChapterTwoStartingScript()
    
    ## Chapter 2 GAME. ZODIAC RIDDLE
    chance = 3 # Total Changce to try
    userPosition = 0 # User position of blocks
    positionStr = 'Starting Point!' # User position in Character
    
    ################ START ZODIAC RIDDLE GAME ###################
    while chance > 0:
        ## Convert UserPostion(Integer) to String positionStr
        if userPosition == 0:
            positionStr = 'Starting Point!'
        elif userPosition == 1:
            positionStr = '1st Block'
        elif userPosition == 2:
            positionStr = '2nd Block'
        elif userPosition == 3:
            positionStr = '3rd Block'
        else:
            positionStr = str(userPosition) + 'th Block'
            
        sprint(f"""
================================================
You have {chance} chances to play!! Let's play!!
Your Position is {positionStr}!!
You have to pass the 12 BLOCKS!!!!
================================================

        """, True)
        input('Press <Enter> to roll the dice!!!!!\n\n')
        sprint("""!!!! You roll the dice!!!!\n""")
        ### ROLL THE DICE
        dice = random.randint(a = 1, b = 6)
        sprint(f"""It is {dice} !!!!!!\n""")
        input('Press <Enter> to Resume.\n\n')
        userPosition = userPosition + dice
        if userPosition < 12: ## Proceeding the Game
            ## Convert UserPostion(Integer) to String positionStr
            if userPosition == 0:
                positionStr = 'Starting Point!'
            elif userPosition == 1:
                positionStr = '1st Block'
            elif userPosition == 2:
                positionStr = '2nd Block'
            elif userPosition == 3:
                positionStr = '3rd Block'
            else:
                positionStr = str(userPosition) + 'th Block'
            sprint(f"""
Now!! you are going {dice} steps!!!

You are on the {positionStr}!!!!

When you step on the block, there is a piece of Riddle!
            
            """, True)
            input('Press <Enter> to Resume.\n\n')
            i = userPosition - 1 # Check User's Current Position as a INDEX
            question = zodiac_cp[i]['question'] ## Pick the question of current position
            zodiac_cpcp = zodiac_cp.copy() ## Copy the question set
            answerList = [] # Answer list
            answerList.append(zodiac_cpcp[i]['name']) # Appending Correct Answer
            zodiac_cpcp.pop(i) # pop the Correct Answer
            random.shuffle(x = zodiac_cpcp) # Shuffle left wrong question lists
            j = 0 
            while j < 3: # Adding 3 other wrong Answers 
                answerList.append(zodiac_cpcp[j]['name']) # Pick random Answer and appending to answer List
                j += 1
            random.shuffle(x = answerList) # Shuffle the answer List
            sprint(f"""
There is a question. 

"{question} What is it?"

Choose an answer below listed by number:

1) {answerList[0]}
2) {answerList[1]}
3) {answerList[2]}
4) {answerList[3]}
            
            """, True)
            result = {
               'msg': 'Message of result', # Response Message
                'win': False # Select Correct Answer is True
            }
            while True:
                answer = input('Please select a number: ')
                if answer != '' and answer in ('1', '2', '3', '4'):
                    answer = int(answer) - 1 # Convert to integer and minus 1
                    if answerList[answer] == zodiac_cp[i]['name']:
                        result['msg'] = 'You are right! It\'s a ' + zodiac_cp[i]['name'] + '!!!!!'
                        result['win'] = True
                        result['status'] = 1
                    else:
                        result['msg'] = 'You are wrong! It\'s a ' + zodiac_cp[i]['name'] + '!!!!!'
                        result['status'] = 1
                    break
                    input('Press <Enter> to Resume.\n\n')
                else:
                    print(""""\n\nPlease select a NUMBER (1 ~ 4).\n\n""")
                    input(prompt='Press <Enter> to Retry. \n')
            ## Print the Result Message    
            sprint('\n')
            sprint(result['msg'])
            sprint('\n')
            input('Press <Enter> to Resume.\n\n')
            ## Result Check
            if not result['win']:
                if result['status'] == 0:
                    print("\nPlease try again!\n")
                else:
                    userPosition = 0
                    chance -= 1
        else:
            response = True
            break
    return response
# End of chapterTwo -------------------------------------------------

# Method Chapter Three
def chapterThree():
    """
    Method Name: chapterThree
    Return: response[boolean] 
    Description: Chapter three - Gussing Who took the most cookies from the cookie jar!
                 Five goblins give users some hint. 
                 User can ask each goblin to recieve hints.
    """
    # START OF CHAPTERTHREE #########################
    # Gat Script
    getChapterThreeStartingScript()
    
    #############################################################
    # WHO TOOK THE MOST COOKIES IN THE COOKIE JAR GAME
    #############################################################
    
    chance = 2
    response = False
    # WHILE START -- Guessing goblin GAME #########################
    while chance > 0:
        print(f"""{chance} Chance(s) left.\n""")
        # The Five Goblin Lists
        goblinList = ['Blue', 'Red', 'Yellow', 'Mint', 'Purple']
        # Goblin ate number of cookies list
        goblinAteCookie = [1, 2, 3, 4, 5]

        # Randomize Goblin list
        goblins = []
        random.shuffle(x = goblinList)
        # Randomize the goblin coopkies for each chance.
        for i, e in enumerate(goblinAteCookie):
            inner = {}
            inner['numCookie'] = e
            inner['name'] = goblinList[i]
            if i == 0: 
                inner['description'] = 'I took 2 fewer cookies than a ' + goblinList[2] + ' Goblin!!'
            elif i == 1: 
                inner['description'] = 'I took two times cookies than a ' + goblinList[0] + ' Goblin!!'
            elif i == 2: 
                inner['description'] = 'I took median cookies'
            elif i == 3:
                inner['description'] = 'I took 2 fewers cookies than a ' + goblinList[4] + ' Goblin!!'
            elif i == 4: 
                inner['description'] = 'I took 2 more cookies than ' + goblinList[2] + ' Goblin!!'
            else:
                print("Unexpected entry. Please check the code for bugs.")
                break
            goblins.append(inner)
        random.shuffle(x = goblins) # Randomize goblin list
        ## Game Result Dictionary
        result =  {
            'msg': 'Message of result', # Response Message
            'win': False # Select Correct Answer is True
        }
        sprint(f"""
They ate Joon's cookies at all!! Joon got angry with them.

Can you guess who ate most cookies?
            
            """, True)
        input('Press <Enter> to Resume.\n\n')
        
        # START WHILE -- First Question Input Control #########################
        while True:
            sprint(f"""
You can ask to the goblins.

1) {goblins[0]['name']} Goblin
2) {goblins[1]['name']} Goblin
3) {goblins[2]['name']} Goblin
4) {goblins[3]['name']} Goblin
5) {goblins[4]['name']} Goblin
--------------------------------
9) Choose the goblin!

            """, True)
            answer = input('Please Select the number to ask goblins. or go forward next stop. \n\n')
            if answer != '' and answer in ['1', '2', '3', '4', '5', '9']:
                answer = int(answer) # convert to integer
                if 1 <= answer <= 5:
                    print(f"""
{goblins[answer - 1]['name']} Goblin Said "{goblins[answer - 1]['description']}"

                    """)
                    input('Press <Enter> to Resume.\n\n')
                elif answer == 9:
                    break
                else:
                    print('\n\nPlease correct number from 1 to 5! Let\'s try again!!')
            else:
                print("Please insert number from option!\n\n")
        # END WHILE -- First Question input Control #########################

        # START WHILE -- Second Question input Control #########################
        while True:
            sprint(f"""
Who ate the most cookies do you think?

1) {goblins[0]['name']} Goblin
2) {goblins[1]['name']} Goblin
3) {goblins[2]['name']} Goblin
4) {goblins[3]['name']} Goblin
5) {goblins[4]['name']} Goblin

        """, True)
            answer = input('Please select the number to pick the goblin who ate the most cookies!!\n\n')
            if answer != '' and answer in ['1', '2', '3', '4', '5']:
                answer = int(answer)
                if 1 <= answer <= 5:
                    if goblins[answer - 1]['numCookie'] == 5:
                        result['win'] = True
                        result['msg'] = 'You are right!!! ' + goblins[answer - 1]['name'] + ' Goblin ate five cookies at all!!!\n'
                    else:
                        result['msg'] = 'You are Wrong!!! ' + goblins[answer - 1]['name'] + ' Goblin did not eat the most number of cookies!!!\n'
                    break
                else:
                    print('\n\nPlease correct number from 1 to 5! Let\'s try again!!')
                    input('Press <Enter> to Resume.\n\n')
            else: 
                print('\n\nPlease insert number from 1 to 5! Let\'s try again!!')
                input('Press <Enter> to Resume.\n\n')
        # END WHILE ###############################
        
        # CHECK THE GAME RESULT
        if result['win']:
            sprint(result['msg'] + '\n')
            input('Press <Enter> to Resume.\n\n')
            response = True
            break
        else:
            sprint(result['msg'] + '\n')
            input('Press <Enter> to Resume.\n\n')
            chance -= 1
    return response
# End of chapterThree -------------------------------------------------
    
# Method chapterFour
def chapterFour():
    """
    Method Name: chapterFour
    Return: response[boolean] 
    Description: Chapter Four
                 Rack, Scissors and paper game with tiger and fox
    """
    # Get Chapter Four Script
    getChapterFourStartingScript()
    
    # Run Rock Scissors Paper game and get Result
    response = rockScissorsPaperResult()
    
    # Return
    return response
# End of chapterFour -------------------------------------------------

# Method rock scissors and papers
def rockScissorsPaperResult():
    """
    Method Name: rockScissorsPaperResult
    Return: response[boolean]
    Description: Run Rock Scissors Paper Game, and return result
                 User can select only Noona's option.
                 Joon, Fox, Tiger select shape themselves
                 If tiger is only winner, the game is over.
                 One of Noona and Joon wins, user will win.
    """
    # Rock Scissors Paper Sets Variable
    rspSets = [
        {'id': 'R', 'name': 'BA-WI'}, # Rock
        {'id': 'S', 'name': 'GA-WI'}, # Scissors
        {'id': 'P', 'name': 'BO'}, # Paper
    ]
    # Players list: 
    players = ['Noona', 'Joon', 'Fox', 'Tiger']
    
    # Declare Response
    response = False
    
    # While Control
    isRun = True
    
    # Round Counter
    round = 1
    
    # Run the Game
    while isRun:
        setList = [] # RSP SETS
        sprint(f"""
Let's play GA-WI BA-WI BO!! (Rock-Scissors-Paper!!)

ROUND {round}!!!

Players are {players} !!!!!!!

        """, True)
        input('Press <Enter> to Resume.\n\n')
        for index, player in enumerate(players):
            if player == 'Noona':
                while True:
                    sprint(f"""
Please select your shape!

                    """)
                    for rsp in rspSets:
                        print(f"""{rsp['id']}: {rsp['name']}""")
                    
                    userSelect = input('Please Select Initial R, S, P (Rock, Sicssiors, Paper)\n\n')
                    if userSelect != '' and userSelect.upper() in ['R','S','P']:
                        for rsp in rspSets:
                            if userSelect.upper() == rsp['id']:
                                setList.append(rsp['id'])
                                break
                            else:
                                pass
                        input('You have selected!\n')
                        break
                    else:
                        print("""Please select correct answer in R, S, P!""")
            else:
                rnd = random.choice(seq = rspSets)
                setList.append(rnd['id'])
                input(f"""{player} has selected!\n""")
        print(f"""
Now!! Let's playing GA-WI BA-WI BO!!!!  
        
        """)
        input('Press <Enter> to Resume.\n\n')
        print(f"""
They are yelling!

GA-WI....(Scissors)

        """)
        input('Press <Enter> to Resume.\n\n')
        print(f"""
BA-WI.......................(Rock)
        
        """)
        input('Press <Enter> to Resume.\n\n')
        print(f"""
BO!!!!!!!!!!!!!!!!!!(Paper!!)
        
        """)
        input('Press <Enter> to Resume.\n\n')
        print("***********************************")
        for l, t in enumerate(setList): # PRINT OUT Result
            print(players[l], end = " has ")
            for rsp in rspSets:
                if t == rsp['id']:
                    print(rsp['name'], end = '!!\n')
                else:
                    pass
        
        print("***********************************")
        input('Press <Enter> to Resume.\n\n')
        setUnique = [] # CHECK UNIQUE RSP
        for e in setList:
            if e not in setUnique:
                setUnique.append(e)
        
        winner = [] # Winner list
        looser = [] # Looser List
        size = len(setUnique) # Count unique shapes
        if size == 2: ## Someone win
            setUnique.sort()
            resultSet = setUnique[0] + setUnique[1]
            if resultSet == 'RS': # Rock win
                for i, e in enumerate(setList):
                    if e == 'R':
                        winner.append(players[i])
                    else:
                        looser.append(players[i])
            elif resultSet == 'PR': # Paper win
                for i, e in enumerate(setList):
                    if e == 'P':
                        winner.append(players[i])
                    else:
                        looser.append(players[i])
            elif resultSet == 'PS': # Scissors win
                for i, e in enumerate(setList):
                    if e == 'S':
                        winner.append(players[i])
                    else:
                        looser.append(players[i])
            else:
                print("Unexpected entry. Please check the code for bugs.")
            sprint(f"""
The winner is {winner}!!!

And the looser is {looser}!!!
            
            """)
            input('Press <Enter> to Resume.\n\n')
            players = winner
            if len(players) == 1 and 'Tiger' in players:
                response = False
                isRun = False
            else:
                if 'Tiger' not in players:
                    response = True
                    isRun = False
                else:
                    # Fox can give Noona or Joon a chance to participation
                    if 'Fox' in players and 'Tiger' in players:
                        if 'Noona' not in players and 'Joon' in players:
                            players.append('Noona')
                        elif 'Joon' not in players and 'Noona' in players:
                            players.append('Joon')
                        else:
                            players.append('Noona')
                    else:
                        pass
        else:
            input('It\'s Tie!!! Let\'s try again!!!\n\n')
        round  += 1
    return response
# End of rockScissorsPaperResult -------------------------------------------------

# method fail
def fail():
    """
    # Method Name: fail
    # Description: Game Over!
    """
    print(f"""

 _______________________________________________________   
/                                                       \\
|   _____ _____ _____ _____    _____ _____ _____ _____   |
|  |   __|  _  |     |   __|  |     |  |  |   __| __  |  |
|  |  |  |     | | | |   __|  |  |  |  |  |   __|    -|  |
|  |_____|__|__|_|_|_|_____|  |_____|\___/|_____|__|__|  |
|                                                        |
\\_______________________________________________________/

                                                
    """)
# End of fail -------------------------------------------------
    
# Method win
def win():
    """
    # Method Name: win
    # Description: Ending Desciption!
    """
    # Win Ending Description 1
    sprint(f"""
After a while, the grandmother opened the bedroom door and told the children as she entered.

Halmoni said, "There you are. Come on and eat!"

Joon said, "Halmoni, where have you BEEN?!"

Noona said, "We were looking all over for you Halmoni!"
    
    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    # Win Ending Description 2
    sprint(f"""
Joon said, "YEAH! We saw a hungry rabbit, lots of hungry giangts, a mean cheating tiger, and a wily bushy tailed fox!"

Noona said, "And Halmoni, this door handle followed us home!"

Joon said, "Oh yeah, did you see a bushy fox that came in?"

The children are saying again and again what they experienced, while they are going to the kitchen.

    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    # Win Ending Description 3
    sprint(f"""
Halmoni, their grandmother, smiled and said to the children, giving them some red bean soup.

"You guys must have been on a very interesting adventure! You must be hungry. Go ahead and eat."

They all sat at the table and ate red bean soup deliciously.

At that time, her body was slightly reflected in the wall mirror. 

There seemed to be a glimpse of nine tails.

    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    # Win Ending Description 4
    sprint("""



THANK YOU FOR PLAYING 'Where's Halmoni!'



    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    # Win Ending Description 5
    print("""
    
        ,----,                                                                  
      ,/   .`|       ,--,                                     ,--.              
    ,`   .'  :     ,--.'|    ,---,.            ,---,.       ,--.'|    ,---,     
  ;    ;     /  ,--,  | :  ,'  .' |          ,'  .' |   ,--,:  : |  .'  .' `\   
.'___,/    ,',---.'|  : ',---.'   |        ,---.'   |,`--.'`|  ' :,---.'     \  
|    :     | |   | : _' ||   |   .'        |   |   .'|   :  :  | ||   |  .`\  | 
;    |.';  ; :   : |.'  |:   :  |-,        :   :  |-,:   |   \ | ::   : |  '  | 
`----'  |  | |   ' '  ; ::   |  ;/|        :   |  ;/||   : '  '; ||   ' '  ;  : 
    '   :  ; '   |  .'. ||   :   .'        |   :   .''   ' ;.    ;'   | ;  .  | 
    |   |  ' |   | :  | '|   |  |-,        |   |  |-,|   | | \   ||   | :  |  ' 
    '   :  | '   : |  : ;'   :  ;/|        '   :  ;/|'   : |  ; .''   : | /  ;  
    ;   |.'  |   | '  ,/ |   |    \        |   |    \|   | '`--'  |   | '` ,/   
    '---'    ;   : ;--'  |   :   .'        |   :   .''   : |      ;   :  .'     
             |   ,/      |   | ,'          |   | ,'  ;   |.'      |   ,.'       
             '---'       `----'            `----'    '---'        '---'         
                                                                                
    
    """)
# End of win -------------------------------------------------
# END OF FUNCTIONAL METHOD ################################################################################## #
    
# START SCRIPT FUNCTION ##################################################################################### #
# get Book Introduction
def bookIntroduce():
    """
    Load the book introduce script
    ---------------------------------------------------------------------------
    Where's Halmoni?

    To Mom and Dad, who opened doors, and to Izzy and Ollie, for whom I hope to open many more.
    And to my wingman, Peter.
    Written by Julie Kim @2017
    -------
    For My lovely adorable Daughter, Rhea.
    Gamified by Yunsik Choung. @2023
    -------
    Kim, Julie J., (2017), Where's Halmoni?, Little Bigfoot, Seattle, WA.
    """
    sprint(f"""
{'*' * 40}

Where's Halmoni?

To Mom and Dad, who opened doors, and to Izzy and Ollie, for whom I hope to open many more.

And to my wingman, Peter.

Written by Julie Kim @2017

-------

For My lovely adorable Daughter, Rhea.

Gamified by Yunsik Choung. @2023

-------

Kim, Julie J., (2017), Where's Halmoni?, Little Bigfoot, Seattle, WA.

{'*' * 40}

    """)
    input(prompt = 'Press <Enter> to Start Game.\n\n')

# get Chapter One Starting Scription
def getChapterOneStartScript():
    """
    Chapter 1 Starting Script
    -----------------------------------------------------------------------
    It was one of the most beautiful days. 
    Halmoni, (grandmother in Korean), was cooking red bean soup, a traditional Korean-style food suitable for her grandchildren.
    Suddenly, some package arrived from Amazon.
    She received the package, and it was a traditional Korean window.
    Halmoni put the elegant door in the bedroom and started making red bean porridge again.
    It was then. Suddenly the window opened and the tiger came into the house and headed to the kitchen.
    The tiger stole the red bean soup pot that Halmoni was making and went back to the window.
    Grandma got angry and chased the tiger!!
    Then she was gone just before her grandchildren came back.
    Noona and Joon are coming home after school!
    Joon said, "Halmoni~~ We are here!!!"
    They opened the main door and came into the foyer.
    """
    # Chapter 1 Description.1
    sprint(f"""
It was one of the most beautiful days. 

Halmoni, (grandmother in Korean), was cooking red bean soup, a traditional Korean-style food suitable for her grandchildren.

Suddenly, some package arrived from Amazon.

She received the package, and it was a traditional Korean window.

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')
    
    # Chapter 1 Description.2
    sprint("""
Halmoni put the elegant door in the bedroom and started making red bean porridge again.

It was then. Suddenly the window opened and the tiger came into the house and headed to the kitchen.

The tiger stole the red bean soup pot that Halmoni was making and went back to the window.

Grandma got angry and chased the tiger!!

Then she was gone just before her grandchildren came back.

    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    # Chapter 1 Description.3
    sprint(f"""
***********************************
*   CHAPTER 1. Where's Halmoni?   *
***********************************

Noona and Joon are coming home after school!

Joon said, "Halmoni~~ We are here!!!"

They opened the main door and came into the foyer.

    """, True)
    input('Press <Enter> to Resume.\n\n')

# get Chapter One Clear Scirpt
def getChapterOneClearSciript():
    """
    Chapter One Clear Script
    ----------------------------------------------------------------------------
    Joon said, "Wow! Noona!! It works. You are so clever!!"
    Suddenly, The Lock is unlocked. And Noona and Joon jump into the door!
    Joon said, "Let's go! Noona!! Let's look around outside!
    """
    # Chapter 1 Clear Description.1
    sprint(f"""
Joon said, "Wow! Noona!! It works. You are so clever!!"

Suddenly, The Lock is unlocked. And Noona and Joon jump into the door!

Joon said, "Let's go! Noona!! Let's look around outside!

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')
    # Chapter 1 Clear Description.2
    sprint(f"""
{'*' * 50}

Congratulation! You cleared Chapter one.

{'*' * 50}

    """, True)
    input(prompt = 'Press <Enter> to Resume Next Chapter...\n\n')
    
# Get Chapter Two Starting Script
def getChapterTwoStartingScript():
    """
    Chapter 2 Starting Script
    ---------------------------------------------------------------------
    Noona and Joon jumped into the mysterious door.
    The outside of their house looks different.
    This place looks like a beautiful picture.
    There are big and fantastic mountains colored green and blue.
    Also, their house looks changed to a very old-style Korean house, Cho-Ga Jip.
    The roof is made of many straws, and some giant yellow gourds are on the top.
    Joon said, "It's outside...Should we keep our slippers on?"
    "Yes, we should. This place looks like the picture on the living room's wall. Where am I?"
    Joon said, "Well, it's awesome!! By the way, Noona!! Here are tracks again!!"
    "Oh! I see. Let's follow the tracks!"
    They are following the tracks of some animal's footprint.
    A while ago, they noticed a tiny little animal following them.
    "Oh! Look Joon! Cute Rabbit!!!"
    Rabbit Growls...
    "It looks hungry. Joon, I'll give her some snacks in your backpack!!"
    Joon said, "Hey, Wait."
    Noona takes the milk chocolates.
    "Here, which one would you like?"
    Rabbit said, "너무 맛있다!!!!"
    Joon said, "What? What did it say?"
    "I think it said it was tasty..."
    Rabbit Likes much of chocolates and candies!
    Joon said, " Since you ate my candy, too, tell me something about these tracks."
    Rabbit is staring at tracks on the road. and...
    Suddenly, the rabbit is rushing to the road with big creaming!!!!!!
    Rabbit said, "어어어어!!!! 오노!!!!!!"
    "Hey!!!! Wait!! Why are you running away!!! Stop!!!"
    May be, rabbit knows about what these tracks from!!!!
    You need to chase the rabbit and ask about these!!!
    In this time! You and Joon are going to 12 blocks of road to chase.
    When you start this chasing game!
    You will roll the dice! And you can go through several steps for the dice number.
    But each block has an East-Asian Zodiac Riddle!
    You must select the correct answer to this Riddle. If you choose the wrong answer, you will slip to the starting point! And lose 1 chance!
    Remember! The chances are only 3 times!!!
    """
    sprint(f"""
************************************
*   CHAPTER 2. Tokki (The Rabbit)  *
************************************

Noona and Joon jumped into the mysterious door.

The outside of their house looks different.

This place looks like a beautiful picture.

There are big and fantastic mountains colored green and blue.

Also, their house looks changed to a very old-style Korean house, Cho-Ga Jip.
    
    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    ## Description of Chapter 2.2
    sprint(f"""
The roof is made of many straws, and some giant yellow gourds are on the top.

Joon said, "It's outside...Should we keep our slippers on?"

"Yes, we should. This place looks like the picture on the living room's wall. Where am I?"

Joon said, "Well, it's awesome!! By the way, Noona!! Here are tracks again!!"

"Oh! I see. Let's follow the tracks!"

    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    ## Description of Chapter 2.3
    sprint(f"""
They are following the tracks of some animal's footprint.

A while ago, they noticed a tiny little animal following them.

"Oh! Look Joon! Cute Rabbit!!!"

Rabbit Growls...

"It looks hungry. Joon, I'll give her some snacks in your backpack!!"

    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    ## Description of Chapter 2.4
    sprint(f"""  
Joon said, "Hey, Wait."

Noona takes the milk chocolates.

"Here, which one would you like?"

Rabbit said, "너무 맛있다!!!!"

Joon said, "What? What did it say?"

    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    ## Description of Chapter 2.5
    sprint(f"""
"I think it said it was tasty..."

Rabbit Likes much of chocolates and candies!

Joon said, " Since you ate my candy, too, tell me something about these tracks."

Rabbit is staring at tracks on the road. and...

Suddenly, the rabbit is rushing to the road with big creaming!!!!!!

Rabbit said, "어어어어!!!! 오노!!!!!!"

"Hey!!!! Wait!! Why are you running away!!! Stop!!!"

    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    ## Description of Chapter 2.6
    sprint(f"""
May be, rabbit knows about what these tracks from!!!!

You need to chase the rabbit and ask about these!!!

In this time! You and Joon are going to 12 blocks of road to chase.

When you start this chasing game!

    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    ## Description of Chapter 2.7
    sprint(f"""
You will roll the dice! And you can go through several steps for the dice number.

But each block has an East-Asian Zodiac Riddle!

You must select the correct answer to this Riddle. If you choose the wrong answer, you will slip to the starting point! And lose 1 chance!

Remember! The chances are only 3 times!!!

    """, True)
    input('Press <Enter> to Resume.\n\n')

# get Chapter Two Clear Scirpt
def getChapterTwoClearSciript():
    """
    Chapter Two Clear Script
    ----------------------------------------------------------------------------
    WOW!!! YOU ARE DONE!! YOU GOT THE RABBIT!!!!!
    Joon said, "Wow! Noona!! You are so fast!!!"
    "Hey! Rabbit! We don't wanna hurt you. We are looking for our Hamoni!"
    Rabbit said, "뭐??? Hamoni???"
    Joon said, "Oh! I think it knows about Halmoni!! Hey! do you know about these tracks?"
    Rabbit said, "오마이갓!! 이빨은 이것보다 아주 날카롭고! 눈은 이렇게 아주 무섭게 찢어졌어 m(ㅡㅂㅡ)m"
    Joon said, "Do you think it's talking about a fox or a raccoon?? Or....maybe an angry giant squirrel?"
    Rabbit said, "AND!!! 엄청 길고 엄청 커!!! 아주 커! 그냥 큰게 아니야. VERY 커!!!! 나보다 아주 많이!!! \(ㅡㅂㅡ)/"
    "I don't know. It's talking way too fast. It's adorable."
    Rabbit said, "Ah!!! 그 성질 고약한....Ho.....HO RHANG EE!!!!"
    "Hey....I think it said TIGER."
    Joon said, "Tiger??!!!!!"
    Rabit said, "Oh!!"
    Rabbit takes off her bundle, then finds something inside.
    Rabit laughed, "Kehehehe"
    Finally, the rabbit gives you something and says, "잘 있어!! Ah, 그리고 Chocolate 잘먹었어!! Ba-bye!!"
    Joon said, "I think it just said 'ba-bye.'"
    "What's this? Is it a back-scratcher, right? Let me try!"
    Noona is trying to scratch Joon's back.
    Joon laughed, "Hahahahahahahahahaahahahaha.....hey..........hahahaha....Stop....hahaha....stop it....hahahahahhaaaahaaa."
    "Yup. There is something funny about this back scratcher."
    Joon is lying down.
    "Hey, wake up. Let's go!"
    """
    # Chapter 2 Clear Description.1
    sprint(f"""
WOW!!! YOU ARE DONE!! YOU GOT THE RABBIT!!!!!

Joon said, "Wow! Noona!! You are so fast!!!"

"Hey! Rabbit! We don't wanna hurt you. We are looking for our Hamoni!"

Rabbit said, "뭐??? Hamoni???"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 2 Clear Description.2
    sprint(f"""
Joon said, "Oh! I think it knows about Halmoni!! Hey! do you know about these tracks?"

Rabbit said, "오마이갓!! 이빨은 이것보다 아주 날카롭고! 눈은 이렇게 아주 무섭게 찢어졌어 m(ㅡㅂㅡ)m"

Joon said, "Do you think it's talking about a fox or a raccoon?? Or....maybe an angry giant squirrel?"

    """, True)
    input('Press <Enter> to Resume.\n\n')

    # Chapter 2 Clear Description.3
    sprint(f"""
Rabbit said, "AND!!! 엄청 길고 엄청 커!!! 아주 커! 그냥 큰게 아니야. VERY 커!!!! 나보다 아주 많이!!! \(ㅡㅂㅡ)/"

"I don't know. It's talking way too fast. It's adorable."

Rabbit said, "Ah!!! 그 성질 고약한....Ho.....HO RHANG EE!!!!"

"Hey....I think it said TIGER."

Joon said, "Tiger??!!!!!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 2 Clear Description.4
    sprint(f"""
Rabit said, "Oh!!"

Rabbit takes off her bundle, then finds something inside.

Rabit laughed, "Kehehehe"

Finally, the rabbit gives you something and says, "잘 있어!! Ah, 그리고 Chocolate 잘먹었어!! Ba-bye!!"

Joon said, "I think it just said 'ba-bye.'"

"What's this? Is it a back-scratcher, right? Let me try!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 2 Clear Description.5
    sprint(f"""
Noona is trying to scratch Joon's back.

Joon laughed, "Hahahahahahahahahaahahahaha.....hey..........hahahaha....Stop....hahaha....stop it....hahahahahhaaaahaaa."

"Yup. There is something funny about this back scratcher."

Joon is lying down.

"Hey, wake up. Let's go!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 2 Clear Description.6
    sprint(f"""
{'*' * 30}

Congratulation! You cleared chapter Two.

{'*' * 30}

    """, True)
    input('Press <Enter> to Resume Next Chapter...\n\n')
    
# Method for chapter Three script
def getChapterThreeStartingScript():
    """
    Chapter 3 Starting Script
    ----------------------------------------------------------------------------------------------
    Noona and Joon are walking in the mountain, which has a beautiful sunset-colored cloud.
    They are chasing the tracks of footprints. And it goes to a mysterious pine grove.
    Joon said, "I'm hungry."
    "Ok. Let's find a place to sit.""
    The kids are exhausted and lots of hunger. They found a little rock and sat on that. 
    They are eating the cookies. 
    'Growl~~~~~~~~~~'
    "Wow, you must be really hungry."
    Joon said, "It's not me."
    "It sounds awfully close..."
    At that time, they got noticed their cookies were gone!
    Both yelled, "You ate all the cookies!"
    "I DID NOT!!!"
    Suddenly, the rock that they are sitting on moves!!
    Joon said, "Oh my.....Noona!!! What's this???"
    "Oh! Is it Dok-kke-bee???? I heard from Halmoni! It's a Dok-kke-bee, the Goblin! They love cookies!!!"
    You met the FIVE GOBLINS!!!!
    They are starving! But they like games. If you win the game, the goblins will give you some presents!
    Can you guess who took the most cookies???
    You can ask each goblin, then they teach you a little hint! Please guess who took the most cookies!!
    REMEMBER, You have only 2 CHANCES!!
    """
    ## Description of Chapter 3.1
    sprint(f"""
*******************************************
*   CHAPTER 3. Dok-kke-bee (The Goblins)  *
*******************************************
    
Noona and Joon are walking in the mountain, which has a beautiful sunset-colored cloud.

They are chasing the tracks of footprints. And it goes to a mysterious pine grove.

Joon said, "I'm hungry."

"Ok. Let's find a place to sit.""
    
    """, True)
    input(prompt = 'Press <Enter> to resume. \n\n')
    
    ## Description of Chapter 3.2
    sprint(f"""
The kids are exhausted and lots of hunger. They found a little rock and sat on that. 

They are eating the cookies. 

'Growl~~~~~~~~~~'

"Wow, you must be really hungry."

Joon said, "It's not me."
    
    """, True)
    input('Press <Enter> to resume.\n\n')
    
    ## Description of Chapter 3.3
    sprint(f"""
'Growl~~~~~~Growl~~~~~~~~~~~~~~~~~~~~~~~~~~~~'

"It sounds awfully close..."

At that time, they got noticed their cookies were gone!

Both yelled, "You ate all the cookies!"

"I DID NOT!!!"
    
    """,True)
    input('Press <Enter> to Resume.\n\n')
    
    ## Description of Chapter 3.4
    sprint(f"""
Suddenly, the rock that they are sitting on moves!!

Joon said, "Oh my.....Noona!!! What's this???"

"Oh! Is it Dok-kke-bee???? I heard from Halmoni! It's a Dok-kke-bee, the Goblin! They love cookies!!!"
    
    """, True)
    input('Press <Enter> to Resume.\n\n')
    
    ## Description of Chapter 3.5
    sprint("""\n\n
You met the FIVE GOBLINS!!!!

They are starving! But they like games. If you win the game, the goblins will give you some presents!

Can you guess who took the most cookies???

You can ask each goblin, then they teach you a little hint! Please guess who took the most cookies!!

REMEMBER, You have only 2 CHANCES!!
    
    """, True)
    input('Press <Enter> to Resume.\n\n')
# get Chapter Three Clear Scirpt
def getChapterThreeClearSciript():
    """
    Chapter Three Clear Script
    ----------------------------------------------------------------------------
    The five goblins are having fun with two children.
    Joon said, "Let's get OUT of here Noona, before we lose ALL our snacks!"
    The five goblins are coming to Joon's backpack!
    Joon said, "I SEE YOU!!!"
    The five goblins are stil hungry but, Joon also stil upset!
    And Joon wants to keep his snacks. At that time, the blue goblin reaches out his hand with something.
    The blue goblin said, "마법 열쇠!"
    It looks like a door handle. 
    Noona said, "Should we trade?"
    Joon is frustrated. but..
    Noona said, "Let it go Joon...."
    Joon screamed, "GAHHHHHHH!!!!"
    Noona takes the Magic door handle, and Joon gives them all his snacks in his backpack.
    Noona said, "How many snacks did you put in there, Joon?"
    Joon said, "Not enough."
    Suddenly, the blue goblin goes to the big colorful rock. 
    And set the magic door handle on that.
    Surprisingly, the rock opens; it looks like a door!! And there are tracks of footprints!!
    Joon said, "Tracks!!"
    Noona said, "It looks like a shortcut!"
    Noona and Joon get into that mysterious door, and the magic door handle follows them.
    The goblin said, "또보자! 잘가! 안녕!"
    Joon said, "Noona! The handle followed us!!"
    Noona said, "It's interesting! C'mon, Let's go!!"
    The children continue their adventure.
    """
    # Chapter 3 Clear Description.1
    sprint(f"""
The five goblins are having fun with two children.

Joon said, "Let's get OUT of here Noona, before we lose ALL our snacks!"

The five goblins are coming to Joon's backpack!

Joon said, "I SEE YOU!!!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Clear Description.2
    sprint(f"""
The five goblins are stil hungry but, Joon also stil upset!

And Joon wants to keep his snacks. At that time, the blue goblin reaches out his hand with something.

The blue goblin said, "마법 열쇠!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Clear Description.3
    sprint(f"""
It looks like a door handle. 

Noona said, "Should we trade?"

Joon is frustrated. but..

Noona said, "Let it go Joon...."

Joon screamed, "GAHHHHHHH!!!!"

Noona takes the Magic door handle, and Joon gives them all his snacks in his backpack.

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Clear Description.4
    sprint(f"""
Noona said, "How many snacks did you put in there, Joon?"

Joon said, "Not enough."

Suddenly, the blue goblin goes to the big colorful rock. 

And set the magic door handle on that.

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Clear Description.5
    sprint(f"""
Surprisingly, the rock opens; it looks like a door!! And there are tracks of footprints!!

Joon said, "Tracks!!"

Noona said, "It looks like a shortcut!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Clear Description.6
    sprint(f"""
Noona and Joon get into that mysterious door, and the magic door handle follows them.

The goblin said, "또보자! 잘가! 안녕!"

Joon said, "Noona! The handle followed us!!"

Noona said, "It's interesting! C'mon, Let's go!!"

The children continue their adventure.

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Clear Description.7
    sprint(f"""
{'*' * 30}

Congratulation! You cleared the chapter 3.

{'*' * 30}

    """, True)
    input(prompt = 'Press <Enter> to Reeume Next Chapter...\n\n')
    
# Method get Chapter Four Starting Script
def getChapterFourStartingScript():
    """
    Chapter 4 Starting Script
    ----------------------------------------------------------------------------------------------
    Noona and Joon are walking and walking on the road.
    'ROAR!!!!!!'
    Joon said, "Sounds like thunder-let's hurry."
    At the end of the road, they face the small valley.
    Joon said, "Noona! That's Halmoni's pot!!!"
    There is a big scary tiger and a bushed nine-tail fox!!! And they have a quarrel!!
    Noona said, "But where's Halmioni?"
    Fox said, "나쁜 호랑이! 이러면 안되지!!"
    At that moment, Joon rushes down to them and yells!
    "That's Halmoni's!!! Give it back!!"
    Noona said, "Joon, come back! STOP!!!!"
    The tiger, fox, and Joon are struggling together!!
    But, the tiger does not lay on the Halmoni's Pot!
    Noona said with anger, "STOP IT! Let's do this FAIRLY!!!"
    Let's do GA-WI, BA-WI, BO
    (Rock, Sicssors, Paper!)
    GA-WI is a Scissors, 
    BA-WI is a Rock,
    BO is a Paper!

    """
    ## Description of Chapter 4.1
    sprint(f"""
*******************************************
*   LAST CHAPTER. Ho-rahng-ee (The Tiger)  *
*******************************************
    
Noona and Joon are walking and walking on the road.

'ROAR!!!!!!'

Joon said, "Sounds like thunder-let's hurry."

At the end of the road, they face the small valley.

Joon said, "Noona! That's Halmoni's pot!!!"

    """, True)
    input(prompt = 'Press <Enter> to resume. \n\n')
    
    ## Description of Chapter 4.2
    sprint(f"""
There is a big scary tiger and a bushed nine-tail fox!!! And they have a quarrel!!

Noona said, "But where's Halmioni?"

Fox said, "나쁜 호랑이! 이러면 안되지!!"

    """, True)
    input(prompt = 'Press <Enter> to resume. \n\n')
    
    ## Description of Chapter 4.3
    sprint(f"""
At that moment, Joon rushes down to them and yells!

"That's Halmoni's!!! Give it back!!"

Noona said, "Joon, come back! STOP!!!!"

    """, True)
    input(prompt = 'Press <Enter> to resume. \n\n')
    
    ## Description of Chapter 4.4
    sprint(f"""
The tiger, fox, and Joon are struggling together!!

But, the tiger does not lay on the Halmoni's Pot!

Noona said with anger, "STOP IT! Let's do this FAIRLY!!!"

    """, True)
    input(prompt = 'Press <Enter> to resume. \n\n')

    
# get Chapter Four Clear Scirpt
def getChapterFourClearSciript():
    """
    Chapter Four Clear Script
    ------------------------------------------------------------------------
    Joon said, "I WIN!!!!"
    Noona yelled, "YAY!! HA HAHAAH"
    Tiger said, "안돼!"
    Fox said, "싸다 싸! HAHAHA HeeHeeHee HoHoHo!"
    You won the Rock Scissors and Paper!!
    But...
    Suddenly, the tiger made a scary face and grabbed the pot tightly.
    And he laughed with a very mean smile.
    Joon said, "Hey! That's CHEATING!!"
    Noona thought the tiger wouldn't return the pot, so she took out her back scratcher and started tickling the tiger.
    The tiger couldn't resist the tickling and missed the pot.
    At that time, the nine-tailed fox quickly took the pot and tried to run away.
    Noona said, "Joon! Follow the bushy-tailed fox!!"
    The tiger got angry and took away her back scratcer and broke it.
    The tiger roared with a big scary face, "ROARRRRRRRRRR!!!!"
    Noona screamed because she was scared, "AAAAHHHHGGGGG!!"
    At that moment!!
    Joon so bravely took the magic handle out quickly and put it on the tiger's stomach.
    Then the tiger stopped moving, and the tiger's body began to open like a door.
    The fox said, "얘들아, 우리 빨리 빨리 가자!!"
    Noona said, "That! That's Halmoni's room!!! Let's go!!!"
    The children were rushing into the door, and closed the door quickly!
    Joon said, "Noona, where's the fox?"
    """
    # Chapter 4 Clear Description.1
    sprint(f"""
Joon said, "I WIN!!!!"

Noona yelled, "YAY!! HA HAHAAH"

Tiger said, "안돼!"

Fox said, "싸다 싸! HAHAHA HeeHeeHee HoHoHo!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Clear Description.2
    sprint(f"""
You won the Rock Scissors and Paper!!

But...

Suddenly, the tiger made a scary face and grabbed the pot tightly.

And he laughed with a very mean smile.

Joon said, "Hey! That's CHEATING!!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Clear Description.3
    sprint(f"""
Noona thought the tiger wouldn't return the pot, so she took out her back scratcher and started tickling the tiger.

The tiger couldn't resist the tickling and missed the pot.

At that time, the nine-tailed fox quickly took the pot and tried to run away.

Noona said, "Joon! Follow the bushy-tailed fox!!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Clear Description.4
    sprint(f"""
The tiger got angry and took away her back scratcer and broke it.

The tiger roared with a big scary face, "ROARRRRRRRRRR!!!!"

Noona screamed because she was scared, "AAAAHHHHGGGGG!!"

At that moment!!

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Clear Description.5
    sprint(f"""
Joon so bravely took the magic handle out quickly and put it on the tiger's stomach.

Then the tiger stopped moving, and the tiger's body began to open like a door.

The fox said, "얘들아, 우리 빨리 빨리 가자!!"

Noona said, "That! That's Halmoni's room!!! Let's go!!!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Clear Description.6
    sprint(f"""
The children were rushing into the door, and closed the door quickly!

Joon said, "Noona, where's the fox?"
.
.
.
    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Clear Description.7
    sprint(f"""
{'*' * 30}

Congratulation! You cleared the Last Chapter.

{'*' * 30}

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')
    

# get ChapterOneFailScript
def getChapterOneFailScript():
    # Chapter 1 Fail Description.1
    sprint(f"""
You failed to guess the password.

Joon said, "Noona! I think it's broken."

Noona said, "So do I. Then where can we look for Halmoni?"

Joon said, "Well, I'm too hungry to look for another place. Noona! Please make some Rah-Myeon. And Let's eat!!"

Noona said, "Ok. Have some Rah-Myeon first!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 1 Fail Description.2
    sprint(f"""
{'*' * 30}

You failed to clear the chapter 1.

The kids cannot find their Halmoni.

{'*' * 30}

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

# get ChapterTwoFailScript
def getChapterTwoFailScript():
    # Chapter 2 Fail Description.1
    sprint(f"""
You failed to chase the  rabbit.

Joon said, "Noona! That rabbit is way too fast!!!"

"So do I. Then where can we look for Halmoni? We are getting lost!!"

Joon said, "I'm scared Noona! I wanna go back home now."

    """, True)
    input(prompt = 'Press <Enter> to resume.\n\n')

    # Chapter 2 Fail Description.2
    sprint(f"""
"Where are we? Halmoni!!!!!"

Joon said, "Halmoni!!!!!!!!!"

Unfortunately, they are getting lost in the beautiful world.
.
.
    """, True)
    input(prompt = 'Press <Enter> to resume.\n\n')

    # Chapter 2 Fail Description.3
    sprint(f"""
"Wake up! Hey, children, Wake up!!!!"

Joon is crying, and Noona is rubbing their eyes...

"WAKE UP!!!! Children!!!"

Both said, "AH!!!! Halmoni??????"

    """, True)
    input(prompt = 'Press <Enter> to resume.\n\n')

    # Chapter 2 Fail Description.4
    sprint(f"""
The children are sleeping in the living room.

Halmoni said, "Wake up! Kids! I already cooked yummy red bean soup! Go to the bathroom and wash your hands!"

"Halmoni!! Where have you been?"

    """, True)
    input(prompt = 'Press <Enter> to resume.\n\n')

    # Chapter 2 Fail Description.5
    sprint(f"""
Joon said, "We were looking for you all around the house and.....and we went weird world...we chased the rabbit...we lost...."

Halmoni said, "What? Hahaha, you had a nightmare! I always said never to sleep on the sofa!! It makes you...."

"NO!!!NO!!! Not that!!! It was real!!! We were chasing big and dirty footprints in your kitchen...!!"

Halmoni said, "Ok, easy, easy, Kids! Anyway, Wash your hands!! And get ready to eat!! Hurry up!! Soup is going cold!"
.
.
Is it really a dream???????

    """, True)
    input(prompt = 'Press <Enter> to resume.\n\n')

    # Chapter 2 Fail Description.6
    sprint(f"""  
{'*' * 30}

You failed to clear the chapter 2.

The kids cannot find their Halmoni.

{'*' * 30}

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

# get ChapterThreeFailScript
def getChapterThreeFailScript():
    # Chapter 3 Fail Description.1
    sprint(f"""
Unfortunately, the children could not select the correct answer.

The goblins are laughing and screaming.

"이야!! 우리가 이겼다!! 이제 너희는 우리랑 계속 놀아야해!!"

Joon said, "Noona! What did they say? Can you understand?"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Fail Description.2
    sprint(f"""
Noona said, "No way...I think they said, play together...."

Joon said, "Play??"

Noona said, "Infinitely...."

Joon said, "What???? No...I wanna go!"

The goblins want to play with them forever.
.
.                
    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Fail Description.3
    sprint(f"""
"Wake up! Hey, children, Wake up!!!!"

Joon is crying, and Noona is rubbing their eyes...

"WAKE UP!!!! Children!!!"

Both said, "AH!!!! Halmoni??????"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Fail Description.4
    sprint(f"""
The children are sleeping in the living room.

Halmoni said, "Wake up! Kids! I already cooked yummy red bean soup! Go to the bathroom and wash your hands!"

"Halmoni!! Where have you been?"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Fail Description.5
    sprint(f"""
Joon said, "We were looking for you all around the house and.....and we went weird world...we chased the rabbit and played with goblins...and....we stuck in...."

Halmoni said, "What? Hahaha, you had a nightmare! I always said never to sleep on the sofa!! It makes you...."

"NO!!!NO!!! Not that!!! It was real!!! We were chasing big and dirty footprints in your kitchen...!!"

Halmoni said, "Ok, easy, easy, Kids! Anyway, Wash your hands!! And get ready to eat!! Hurry up!! Soup is going cold!"
.
.
Is it really a dream???????

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 3 Fail Description.6
    sprint(f"""
{'*' * 30}

You failed to clear the chapter Tree.

The kids cannot find their Halmoni.

{'*' * 30}

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')
# get ChapterFourFailScript
def getChapterFourFailScript():
    # Chapter 4 Fail Description.1
    sprint(f"""
Joon said, "Oh! No!!!! We lost."

The tiger laughed with a very mean smile.

The nine-tailed fox held onto the tiger's pot and spoke in poor English!

"Go! 아가들, Just go! Run, 도망가, Run away!!! Hurry!!!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Fail Description.2
    sprint(f"""
Noona said, "Joon! Let's get out away from here!!"

Joon said, "But....We still can not find Halmoni!!"

Noona said, "Just go! C'mon! We have to get out of here!!!"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Fail Description.3
    sprint(f"""
Leaving the tiger and the fox fighting over the pot, the children rushed away.

After running away for a long time, they wandered through the beautiful forest again.

Children are too tired to walk around more. 

Joon said, "Noona, I'm so sleepy...and hungry..."

Noona said, "So do I...where can we go back home?"

    """, True)
    input(prompt = 'Press <Enter> to resume.\n\n')

    # Chapter 4 Fail Description.4
    sprint(f"""
"Wake up! Hey, children, Wake up!!!!"

Joon is crying, and Noona is rubbing their eyes...

"WAKE UP!!!! Children!!!"

Both said, "AH!!!! Halmoni??????"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Fail Description.5
    sprint(f"""
The children are sleeping in the living room.

Halmoni said, "Wake up! Kids! I already cooked yummy red bean soup! Go to bathroom and wash your hands!"

"Halmoni!! Where have you been?"

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Fail Description.6
    sprint(f"""
Joon said, "We were looking for you all around he house and.....and we went weird world...we fought with big scary tiger...and....we stuck in...."

Halmoni said, "What? Hahaha, you had a nightmare! I always said never to sleep on the sofa!! It makes you...."

"NO!!!NO!!! Not that!!! It was real!!! We were chasing big and dirty footprints in your kitchen...!!"

Halmoni said, "Ok, easy, easy, Kids! Anyway, Wash your hands!! And get ready to eat!! Hurry up!! Soup is going cold!"
.
.
Is it really a dream???????

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

    # Chapter 4 Fail Description.7
    sprint(f"""
{'*' * 30}

You failed to clear the Last Chapter.

The kids cannot find their Halmoni.

{'*' * 30}

    """, True)
    input(prompt = 'Press <Enter> to Resume.\n\n')

# END SCRIPT FUNCTION ##################################################################################### #    


# Call startGame()
startGame()
```
